```python
from gingerit.gingerit import GingerIt
```

```python
data = []
corrector = GingerIt()
with open('/content/drive/MyDrive/Colab Notebooks/RELPRON/RELPRON/relpron.all', 'r') as file:
  for line in file:
    line = line.split()
    instance = {}
    instance['class'] = line[0]
    if line[0] == "OBJ":
      sentence = ['A'] + [line[1]] + ['is'] + ["a"] + line[2:4] + ["a"] + line[4:] + ['.']
    else:
      sentence = ['A'] + [line[1]] + ['is'] + ["a"] + line[2:5] + ['a'] + line[5:] + ['.']

    new_sentence= []
    for word in sentence:
      if "_" in word:
        word = word.split('_')[0]
      new_sentence.append(word)
    if line[0] == "OBJ":
      if new_sentence[-2][-1] != "s":
        new_sentence[-2] +="s"
      correct = corrector.parse(' '.join(new_sentence))
      instance['sentence'] = correct["result"]
      #instance['sentence'] = ' '.join(new_sentence)
    if line[0] == "SBJ":
      correct = corrector.parse(' '.join(new_sentence))
      instance['sentence'] = correct["result"]

    print(instance['sentence'])
    data.append(instance)
```

```python
sannotated_data = []
for instance in data:
  if instance['class'] == "OBJ":
    instance['pred_arg'] = (9,5)
    instance['label'] = "Patient"
    annotated_data.append(instance.copy())

    instance['pred_arg'] = (9,8)   
    instance['label'] = "Agent"
    annotated_data.append(instance.copy())
  else:
    instance['pred_arg'] = (7,5)
    instance['label'] = "Agent"
    annotated_data.append(instance.copy())

    instance['pred_arg'] = (7,9)
    instance['label'] = "Patient"
    annotated_data.append(instance.copy())
```

```python
print(annotated_data)
```
